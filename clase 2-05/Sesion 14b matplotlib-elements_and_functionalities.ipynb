Para incrustar una leyenda en un gráfico hay que incluir la linea `plt.legend`

```
from matplotlib import pyplot as plt
%matplotlib inline

AA_data = data[data["stock"]=="AA"]
plt.plot(AA_data.date,AA_data.high, label ="high", c="green")
plt.plot(AA_data.date,AA_data.low, label ="low", c="red")
plt.legend(loc="best")
plt.show()
```

```
main = AA_data[["low","high"]].mean(axis=1)
plt.plot(AA_data.date, AA_data.high, label ="high", c="green")
plt.plot(AA_data.date, AA_data.low, label ="low", c="red")
# This function creates a scatter plot with date type variables
plt.plot_date(AA_data.date, main, marker="+",label="main")
plt.legend(loc="best")
plt.show()
```

# Anotaciones



Para incluir anotaciones se utiliza `plt.annotate`. Las anotaciones están basadas en flechas o relaciones de una coordenada con un texto. Como ejemplo indicaremos el minimo de los valores "low" y el maximo de los valores "max" con anotaciones.

```
AA_high_max = AA_data["high"].max()
AA_high_max_date = AA_data.loc[AA_data["high"].idxmax(),"date"]
```

```
AA_low_min = AA_data["low"].min()
AA_low_min_date = AA_data.loc[AA_data["low"].idxmin(),"date"]
```

```
import matplotlib.dates as mdates

# Add a black arrow annotation
plt.annotate("Max - High {}".format(AA_high_max),
    xy=(AA_high_max_date,AA_high_max),
    xytext=(mdates.date2num(AA_high_max_date)+5, AA_high_max+.5),
    arrowprops=dict(facecolor='black'))

plt.annotate("Min - Low {}".format(AA_low_min),
    xy=(AA_low_min_date,AA_low_min),
    xytext=(mdates.date2num(AA_low_min_date)-5, AA_low_min+.5),
    arrowprops=dict(facecolor='black'))

plt.title("AA Stock values")
plt.plot(AA_data.date, AA_data.high, label ="high", c="green")
plt.plot(AA_data.date, AA_data.low, label ="low", c="red")
plt.show()
```



https://matplotlib.org/users/annotations.html

```
def custom_plot():
    plt.annotate("Max - High {}".format(AA_high_max),
    xy=(AA_high_max_date,AA_high_max),
    xytext=(mdates.date2num(AA_high_max_date)+5, AA_high_max+.5),
    arrowprops=dict(facecolor='black'))

    plt.annotate("Min - Low {}".format(AA_low_min),
    xy=(AA_low_min_date,AA_low_min),
    xytext=(mdates.date2num(AA_low_min_date)-5, AA_low_min+.5),
    arrowprops=dict(facecolor='black'))

    plt.title("AA Stock values")
    plt.plot(AA_data.date, AA_data.high, label ="high", c="green")
    plt.plot(AA_data.date, AA_data.low, label ="low", c="red")
    plt.show()
```
