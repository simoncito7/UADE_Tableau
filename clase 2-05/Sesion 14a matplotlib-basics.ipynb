## Histograma: Una variable continua

Los histogramas representan en altura la acumulación de casos en intervalos de valor. Una sola variable continua puede representarse de esta manera.

```
plt.hist(data_gaussian_1, bins=50);
plt.show()
```



Para el ejemplo, crearemos varias distribuciones de datos, y observaremos como se muestran en histogramas.

```
data_gaussian_2 = np.random.normal(loc=1.5, scale=0.1, size=500)
data_gaussian_3 = np.random.normal(loc=0.5, scale=0.6, size=500)
```

Y las graficamos

```
plt.hist(data_gaussian_2, bins=20, label='data 1')
plt.hist(data_gaussian_3, bins=20, label='data 2')

plt.legend(loc='best');
plt.show()
```



Pero se ve extraño, ¿porqué pasa?

```
bins = np.linspace(-3, 3, 50)

# Set an alpha and use the same bins since we are plotting two hists
plt.hist(data_gaussian_2, bins=bins, alpha=0.5, label='data 1')
plt.hist(data_gaussian_3, bins=bins, alpha=0.5, label='data 2')

plt.legend(loc='best');
plt.show()
```



## Barras: Una variable categórica

Para representar una variable categórica usamos el clásico diagrama de barras. Se representa en alturas (o longitudes) ciertas cantidades asociadas a ciertos valores.



```
cities = [
"New York",
"Chicago",
"San Francisco",
"Los Angeles",
"Montgomery"]

numbers = [np.random.randint(10) for city in cities]
positions = list(range(len(cities)))
print(cities)
print(positions)
print(numbers)
```

```
plt.bar(positions, numbers, align="center",tick_label=cities)
plt.show()
```

```
plt.barh(positions, numbers, align="center",tick_label=cities)
plt.show()
```

## Scatter (variables numéricas sin relacion)

```
data_gaussian_4 = np.random.normal(loc=0.5, scale=2, size=500)
plt.scatter(data_gaussian_2, data_gaussian_3, alpha=0.5);
plt.show()
```



Para reprepresentar varios conjuntos de variables, por defecto, cambia los colores según un esquema propio.

Pero se pueden especificar diferencias especificas:

`marker` para la forma del item -> https://matplotlib.org/api/markers_api.html
`color` para el color
`size` para el tamaño
...

```
plt.scatter(data_gaussian_2, data_gaussian_3, alpha=0.5, marker="*");
plt.scatter(data_gaussian_2, data_gaussian_4, alpha=0.3, marker="8");


plt.show()
```



Tambien podemos combinar varias tipos en una grafica. Pero hay que tener cuidado con los ejes.

```
sizes = np.round(np.abs(data_gaussian_4)*20,0)
sizeb = np.round((np.abs(data_gaussian_2)-1)*150,0)
plt.scatter(data_gaussian_2, data_gaussian_4, alpha=0.5, marker="8", s=sizes);
plt.scatter(data_gaussian_2, data_gaussian_3, alpha=0.5, marker="*", s=sizeb);

plt.show()
```

```
data_gaussian_4 = np.random.normal(loc=0.5, scale=2, size=500)
plt.scatter(data_gaussian_2, data_gaussian_3, alpha=0.5, marker="*");
plt.scatter(data_gaussian_2, data_gaussian_4, alpha=0.2, marker="8");

plt.hist(data_gaussian_2, bins=bins, alpha=0.5, label='data 1')

plt.show()
```

También se puede unir la secuencia de puntos, con lineas mediante un plot



A la que se le pueden añadir otros elementos: Varias lineas, leyenda, valores ... 

```
x = np.linspace(0, 2, 10)

plt.plot(x, x, 'o-', label='Linear') # modify type of representation and assigned label 
plt.plot(x, x ** 2, 'x-', label='Quadratic')

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Linear and quadratic relationship')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');

plt.show()
```

Gráfico la función Seno

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.show()
```

Agrego título y leyenda

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Seno de X')  # main title

plt.show()
```

Agrego etiquetas a los ejes

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Seno de X')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');


plt.show()
```

Agrego una grilla

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Seno de X')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');
plt.grid(True)
plt.axis([0, 2*np.pi, -1, 1])




plt.show()
```

Cambio la posición de la grilla

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Seno de X')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');
plt.grid(True)
plt.axis([0, 2*np.pi, -1, 1])

plt.xticks([ 0, np.pi/2, np.pi,3*np.pi/2,2*np.pi])
plt.yticks([-1, 0, +1])

plt.show()
```

Cambio las etiquetas de los puntos de la grilla

```
plt.rc('figure', figsize=(10, 5))
x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x),  label='sin(x)') # modify type of representation and assigned label

plt.legend(loc='best') # indicate the parameter in order to show the legend

plt.title('Seno de X')  # main title

plt.xlabel('Independent axis')
plt.ylabel('Dependent axis');
plt.grid(True)
plt.axis([0, 2*np.pi, -1, 1])

plt.xticks([ 0, np.pi/2, np.pi,3*np.pi/2,2*np.pi])
plt.yticks([-1, 0, +1])

plt.xticks([ 0, np.pi/2, np.pi,3*np.pi/2,2*np.pi],
          [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])

plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])


plt.show()
```
